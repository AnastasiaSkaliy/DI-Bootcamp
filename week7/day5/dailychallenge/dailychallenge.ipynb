Data Cleaning And Preprocessing:

```python
missing_values = df.isnull().sum()
print(missing_values)
```

Missing data in columns 'approxdate' and 'realted' are detected. We have to explore these columns to check how to handle missing info.

For a reson that we have a year without missing data, and we have a column approximate date with 95% of missing data, I decided to drop the column 'approxdate'.

```python
df.drop('approxdate', axis=1, inplace=True)
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.dtypes
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:")
print(categorical_columns)
```

```python
# Specify columns to convert to one-hot encoding
columns_to_one_hot = ['multiple', 'crit1', 'crit2', 'crit3', 'success', 'suicide']

df_encoded = pd.get_dummies(df, columns=columns_to_one_hot)
```

We'll check later what columns we neet also to convert to one-hot encoding, according to what we'll analyze.

```python
#Change month 0 to most frequently month
most_frequent_month = df['imonth'].mode().iloc[0]
df['imonth'] = df['imonth'].replace(0, most_frequent_month)
```

Numerical Normalization:

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

columns_to_normalize = ['INT_LOG', 'INT_IDEO', 'INT_MISC', 'INT_ANY']

# Min-Max scaling
min_max_scaler = MinMaxScaler()
df[columns_to_normalize] = min_max_scaler.fit_transform(df[columns_to_normalize])

# Z-score standardization
standard_scaler = StandardScaler()
df[columns_to_normalize] = standard_scaler.fit_transform(df[columns_to_normalize])
```

Let's analyze attack frequency by country 

```python
import matplotlib.pyplot as plt
import seaborn as sns

attacks_by_country = df['country_txt'].value_counts()
top_30_countries = attacks_by_country.head(30)

plt.figure(figsize=(15, 8))
sns.barplot(x=top_30_countries.index, y=top_30_countries.values, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.title('Top 30 Countries by Attack Frequency')
plt.show()
```

Let's explore the number of attacks iver the years

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.countplot(x='iyear', data=df, palette='viridis')
plt.title('Number of Attacks Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
import matplotlib.pyplot as plt

weapons_columns = ['weaptype1', 'weaptype2', 'weaptype3']

weapons_data = pd.concat([df[col].dropna() for col in weapons_columns])

weapons_counts = weapons_data.value_counts()

plt.figure(figsize=(8, 6))
plt.pie(weapons_counts, labels=weapons_counts.index, autopct='%1.1f%%', startangle=140, colors =sns.color_palette("pastel") )
plt.title('Distribution of Weapon Types')
plt.show()
```

●	Analyzing danger levels in a region for tourism.

```python
attacks_by_month = df['imonth'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=attacks_by_month.index, y=attacks_by_month.values, marker='o', color='purple')
plt.title('Number of Attacks Over the Months')
plt.xlabel('Month')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.xticks(range(1, 13))
plt.show()
```

We can observe that the highest number of attacks occurs during the summer season, with the peak in May.

○	Analyze frequency to give a danger level (success) to all regions.

```python
attacks_by_region_success = df.groupby(['region_txt', 'success']).size().unstack(fill_value=0)

plt.figure(figsize=(16, 8))
attacks_by_region_success.plot(kind='bar', stacked=True, colormap="viridis")
plt.title('Frequency of Attacks by Success Level and Country')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.legend(title='Success Level', labels=['Unsuccessful (0)', 'Successful (1)'], loc='upper right')
plt.show()
```
