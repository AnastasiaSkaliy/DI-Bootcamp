```python
import numpy as np
import pandas as pd
np.random.seed(42)
num_samples = 1000
FeatureA = np.random.rand(num_samples) * 1000
FeatureB = np.random.rand(num_samples) * 10

print("FeatureA:", FeatureA[:5]) #see first 5 
print("FeatureB:", FeatureB[:5])
data = np.column_stack((FeatureA, FeatureB))
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Apply Min-Max scaling
df_normalized = scaler.fit_transform(data)

print("Normalized Data:")
print(df_normalized)
```

```python
import matplotlib.pyplot as plt
# Separate scaled features
scaled_FeatureA = df_normalized[:, 0]
scaled_FeatureB = df_normalized[:, 1]

plt.figure(figsize=(8, 6))
plt.scatter(scaled_FeatureA, scaled_FeatureB, alpha=0.6)
plt.title('Scaled Data: FeatureA vs FeatureB')
plt.xlabel('FeatureA (Scaled)')
plt.ylabel('FeatureB (Scaled)')
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Apply Z-score standardization
df_standardized = scaler.fit_transform(data)
print("Standardized Data:")
print(df_standardized)
```

```python
# Separate scaled features
standardized_FeatureA = df_standardized[:, 0]
standardized_FeatureB = df_standardized[:, 1]

plt.figure(figsize=(8, 6))
plt.scatter(standardized_FeatureA, standardized_FeatureB, alpha=0.6)
plt.title('Standardized Data: FeatureA vs FeatureB')
plt.xlabel('FeatureA (Standardized)')
plt.ylabel('FeatureB (Standardized)')
plt.grid(True)
plt.show()
```

In the case of the synthetic dataset with FeatureA in the range [0, 1000] and FeatureB in the range [0, 10]:

Min-Max Scaling would keep the original range intact for both features, transforming them to a range of [0, 1]. This might be suitable if preserving the original range is essential for interpretation or specific algorithms that require values within that range.
Z-score Normalization would standardize the features to have mean 0 and standard deviation 1. This transformation might be more suitable if the distribution of FeatureA and FeatureB varies significantly and when the relative differences between data points are more relevant than their specific ranges.
Based on the distribution and range of the data, if the goal is to focus on the relative relationships between the data points while minimizing the influence of outliers, Z-score Normalization might be more appropriate. However, if preserving the original range of values is crucial or if the features are inherently bounded within certain limits, Min-Max Scaling could be preferred.

```python
min_range_A, max_range_A = 1, 100
min_range_B, max_range_B = 1000, 10000

num_samples = 100
FeatureA = np.random.uniform(min_range_A, max_range_A, num_samples)
FeatureB = np.random.uniform(min_range_B, max_range_B, num_samples)

print("FeatureA:", FeatureA[:5]) #see first 5 
print("FeatureB:", FeatureB[:5])

data = np.column_stack((FeatureA, FeatureB))
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Apply Min-Max scaling
df_normalized = scaler.fit_transform(data)

print("Normalized Data:")
print(df_normalized)
```

```python
#Plot the original dataset
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)

plt.scatter(FeatureA, FeatureB, alpha=0.6)
plt.title('Original Data: FeatureA vs FeatureB')
plt.xlabel('FeatureA (Original)')
plt.ylabel('FeatureB (Original)')

# Plot the scaled dataset

scaled_FeatureA = df_normalized[:, 0]
scaled_FeatureB = df_normalized[:, 1]
plt.subplot(1, 2, 2)
plt.scatter(scaled_FeatureA, scaled_FeatureB, alpha=0.6)
plt.title('Scaled Data: FeatureA vs FeatureB')
plt.xlabel('FeatureA (Scaled)')
plt.ylabel('FeatureB (Scaled)')

plt.tight_layout()
plt.show()
```

```python
# Plot the scaled dataset

scaled_FeatureA = df_normalized[:, 0]
scaled_FeatureB = df_normalized[:, 1]

plt.figure(figsize=(10, 5))
plt.scatter(scaled_FeatureA, scaled_FeatureB, alpha=0.6)
plt.title('Scaled Data: FeatureA vs FeatureB')
plt.xlabel('FeatureA (Scaled)')
plt.ylabel('FeatureB (Scaled)')
plt.grid(True)
plt.show()
```

Min-Max Scaling transformed FeatureA and FeatureB from their original ranges ([1, 100] and [1,000, 10,000] respectively) to a common range of [0, 1]. This scaling ensures that both features now share a standardized range, simplifying comparisons and preventing dominance by larger magnitude values.

```python
mean_X, std_dev_X = 50, 10
mean_Y, std_dev_Y = 500, 100
num_samples = 1000

# Random data with normal distributions
FeatureX = np.random.normal(mean_X, std_dev_X, num_samples)
FeatureY = np.random.normal(mean_Y, std_dev_Y, num_samples)

print("FeatureX:", FeatureX[:5])
print("FeatureY:", FeatureY[:5])

data = np.column_stack((FeatureX, FeatureY))
```

```python
#Apply Z-score Normalization
df_standardized = scaler.fit_transform(data)
print("Standardized Data:")
print(df_standardized)
```

```python
plt.figure(figsize=(10, 5))

# Box plot for the original dataset
plt.subplot(1, 2, 1)
plt.boxplot(data, labels=['FeatureX', 'FeatureY'])
plt.title('Original Dataset')
plt.ylabel('Values')

# Box plot for the scaled dataset
plt.subplot(1, 2, 2)
plt.boxplot(df_standardized, labels=['FeatureX', 'FeatureY'])
plt.title('Z-score Normalized Dataset')
plt.ylabel('Standardized Values')

plt.tight_layout()
plt.show()
```

Z-score Normalization shifts the mean of each feature to 0 and sets their standard deviation to 1. For instance, FeatureX and FeatureY, initially with means of 50 and 500, and standard deviations of 10 and 100 respectively, are transformed to have means of 0 and standard deviations of 1 after normalization. This process centers the data around 0 and scales it uniformly, aiding in comparing and analyzing features with different scales.

```python
df.columns
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Defining bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtering the outliers
outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
print(outliers)
```

```python
# Trim outliers
trimmed_df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
#Capping
from scipy.stats.mstats import winsorize

# Winsorize data
winsorized_data = winsorize(df['price'], limits=[0.05, 0.05])  # Capping the bottom and top 5%

# Convert to DataFrame for further analysis if needed
winsorized_df = pd.DataFrame(winsorized_data, columns=['price'])
```

```python
# Log transformation
df['price'] = np.log(df['price'] + 1)

# Square root transformation
df['price'] = np.sqrt(df['price'])

# Box-Cox transformation
from scipy import stats
df['price'], _ = stats.boxcox(df['price'] + 1)  # Adding 1 to avoid issues with 0 values
```

Pros and Cons:
Trimming:
Pros: Preserves the original data structure.
Cons: Loss of data; might remove important information, especially if outliers contain valuable insights.
Capping:
Pros: Retains all data points; helps in controlling extreme values.
Cons: Alters the original data, potentially affecting statistical analysis and model performance if outliers are significant.
Transformation:
Pros: Reduces the impact of outliers without losing data.
Cons: Alters the interpretation of the 'price' variable; requires reverse transformation for meaningful interpretation.
Choosing the appropriate technique depends on the context and the importance of outliers in the analysis. For instance, in financial analysis, preserving outliers might be crucial for risk assessment, while in modeling, capping or transformation might improve model performance without losing too much information.

```python
df = pd.read_csv(r'/Users/admin/Desktop/python/di-bootcamp/week7/day3/WineQT.csv')
df.head()
```

```python
columns_of_interest = ['fixed acidity', 'residual sugar', 'alcohol']
selected_data = df[columns_of_interest]

z_scores = np.abs(stats.zscore(selected_data))
outliers = np.where((z_scores > 3) | (z_scores < -3))
```

```python
# Remove outliers from the dataset
cleaned_data = selected_data[(z_scores < 3).all(axis=1)]

# Calculate mean and standard deviation with outliers
stats_with_outliers = selected_data.agg(['mean', 'std'])

# Calculate mean and standard deviation without outliers
stats_without_outliers = cleaned_data.agg(['mean', 'std'])

# Display statistics
print("Statistics WITH outliers:")
print(stats_with_outliers)

print("\nStatistics WITHOUT outliers:")
print(stats_without_outliers)
```


Outliers, the extreme values in data, can distort our understanding of wine features like acidity, sugar content, and alcohol level. They skew averages, inflate variability, and can misrepresent the true distribution of values. In wine quality assessment, outliers might highlight exceptional characteristics or indicate flaws that affect overall judgment. Ignoring outliers risks inaccurate assessments, affects decision-making, and can mislead predictive models. Considering outliers carefully ensures a more accurate interpretation of wine quality and any data analysis, leading to better-informed decisions and more reliable insights.
