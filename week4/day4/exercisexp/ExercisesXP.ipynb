```python
import pandas as pd
df = pd.read_csv('train.csv')
df.head()
```

```python
len(df) #Display the number of rows before removing duplicates
```

```python
#Ex.1
duplicate_rows = df[df.duplicated()] # Identify if there are any duplicate rows
```

```python
print(duplicate_rows) #print duplicates rows
```

```python
df = df.drop_duplicates() # remove dupicates rows
```

```python
print(df.duplicated().sum()) #Verify duplicates are removed
```

```python
len(df) #Display the number of rows after removing duplicates
```

```python
#Ex.2
missing_data = df.isnull() # Check for missing values in a DataFrame
print(missing_data.head())
```

```python
missing_data = df.columns[df.isnull().any()].tolist() #Another method - column Age, Cabin, Embarked  have missing volues
print(missing_data)
```

```python
#Explore different strategies for handling missing data, such as removal, imputation, and filling with a constant value.
df.dropna(subset=['Age'], inplace=True)
```

```python
na_variables = [ var for var in df.columns if df[var].isnull().mean() > 0 ]
data_na = df[na_variables].isnull().mean()
```

```python
#imputation
import numpy as np
from sklearn.impute import SimpleImputer
df['Cabin']= SimpleImputer(missing_values=np.nan, strategy='mean')
df.isnull().sum()
```

```python
imputer = SimpleImputer(strategy='median')
df['Age'] = imputer.fit_transform(df[['Age']])
```

```python
df['Embarked'].fillna('Unknown', inplace=True) # Filling with a constant value
print(df)
```

```python
missing_data = df.isnull() # Check for missing values in a DataFrame - NO MISSING DATA LEFT
print(missing_data.head())
```

```python
missing_data = df.columns[df.isnull().any()].tolist() #Another method - NO MISSING DATA LEFT
print(missing_data)
```

```python
#Ex.3
df['Family Size'] = df['SibSp'] + df['Parch']
```

```python
df['Title'] = df['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
df[['Name', 'Family Size', 'Title']].head()

```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
#one-hot encoding
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked')
df = pd.concat([df, embarked_dummies], axis=1)
df[['Embarked']].head()
```

```python
#Normalize or standardize numerical features if required.
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
scaler = StandardScaler()
df['Age'] = scaler.fit_transform(df['Age'].values.reshape(-1, 1))
df.head()
```
