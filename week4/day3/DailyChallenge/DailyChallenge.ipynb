```python
#Data Loading
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
#Export to Excel
output_file = 'cleaned_iris.xlsx'
df.to_excel(output_file, index=False)
```

```python
#Import from Excel
excel_file = 'cleaned_iris.xlsx'
df_from_excel = pd.read_excel(excel_file)
print(df_from_excel.head())
```

```python
#Save Subset to JSON
subset = df_from_excel.loc[:9, ['sepal_length', 'species']]  # Selecting first 10 rows and 'sepal_length', 'species' columns
json_file = 'iris_subset.json'
subset.to_json(json_file, orient='records')
```
