```python
df = df[df['population'] < 10000]
```

```python
df['population'].hist(bins=50)
```

```python
df.mean()
```

```python
df['avg_family_size'] = df['population'].div(df['households'])
```

```python
df['avg_family_size'].head()
```

```python
df = df[df['avg_family_size'] <= 10]
fig, ax = plt.subplots()
ax.scatter(df['avg_family_size'], df['median_house_value'])
ax.set(title='Median House Value by Avg Family size', xlabel='avg family size', ylabel='median house value')
```
