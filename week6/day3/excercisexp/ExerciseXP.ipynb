```python
import numpy as np
import scipy.stats as stats

data = [12, 15, 13, 12, 18, 20, 22, 21]
mean_scipy = stats.tmean(data)
median_scipy = np.median(data)  # SciPy does not have a separate median function
variance_scipy = stats.tvar(data)
std_dev_scipy = stats.tstd(data)

print(mean_scipy)
print(median_scipy)
print(variance_scipy)
print(std_dev_scipy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics

# Calculating mean and standard deviation
mean = 50
std_dev = 10
data = np.random.normal(mean, std_dev, 1000)

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

pdf = norm.pdf(x, mean, std_dev)

# Plot the histogram of the generated data
plt.hist(data, bins=30, density=True, alpha=0.5, label='Generated Data')

# Plot the probability density function (PDF) of the normal distribution
plt.plot(x, pdf, label='Normal Distribution PDF', color='red')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution with mean=50, std_dev=10')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import ttest_ind

np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=55, scale=12, size=100)

#t test
t_statistic, p_value = ttest_ind(data1, data2)


print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the datasets.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the datasets.")
```

```python
from scipy.stats import skew, kurtosis

data = np.random.normal(0, 1, 1000)

skewness = skew(data)
kurt = kurtosis(data)

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
```
