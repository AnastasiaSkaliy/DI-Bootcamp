```python
import hashlib

def anonymize_species(df):
    # Creating a new column 'AnonymizedSpecies' using SHA-256 hash of the 'Species' column
    df['AnonymizedSpecies'] = df['Species'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest())
    return df

# Anonymize the 'Species' column
df_anonymized = anonymize_species(df)
df_anonymized.head()
```
