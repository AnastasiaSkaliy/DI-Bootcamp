```python
# Convert 'Order Date' column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
```

```python
# Filter data for the years 2018 to 2020
filtered_df = df[(df['Order Date'].dt.year >= 2018) & (df['Order Date'].dt.year <= 2020)]
```

```python
# Group by 'Category' and 'Sub-Category'
grouped_df = filtered_df.groupby(['Category', 'Sub-Category'])
```

```python
# Calculate total sales and average discount for each group
result_df = grouped_df.agg({'Sales': 'sum'}).reset_index()
print(result_df)
```

```python
#Sort the results and visualize the top 5 sub-categories in each category by sales.
sorted_df = result_df.sort_values(by='Sales', ascending=False)
print(sorted_df)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=len(sorted_df['Category'].unique()), ncols=1, figsize=(10, 8))
for i, (category, group) in enumerate(sorted_df.groupby('Category')):
    top5_subcategories = group.head(5)
    ax = axes[i]
    top5_subcategories.plot(kind='bar', x='Sub-Category', y='Sales', ax=ax, legend=False)
    ax.set_title(category)
    ax.set_xlabel('Sub-Category')
    ax.set_ylabel('Sales')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'/Users/admin/Desktop/python/di-bootcamp/week9/day2/exercisexp/train.csv')
df.head()
```

```python
#Create a new DataFrame summarizing customer-level metrics
customer_df = df.groupby('Customer ID').agg({
    'Order ID': 'nunique',   # Purchase frequency
    'Sales': ['mean', 'sum'] # Average purchase value and total sales
}).reset_index()
```

```python
customer_df.head()
```

```python
customer_df.columns = ['Customer ID', 'Purchase Frequency', 'Average Purchase Value', 'Sales']
```

```python
#Calculate CLV for each customer
# A time period (for example 3 years) for CLV calculation
clv_period = 3
customer_df['CLV'] = (customer_df['Average Purchase Value'] * customer_df['Purchase Frequency']) * clv_period
```

```python
#Identify top customers by CLV and visualize their contribution.
top_customers = customer_df.sort_values(by='CLV', ascending=False).head(10) #top 10 customers 
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Customer ID', y='CLV', data=top_customers)
plt.title('Top Customers by CLV')
plt.xlabel('Customer ID')
plt.ylabel('Customer Lifetime Value (CLV)')
plt.show()
```

ChatGPT answer:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your original DataFrame
# If you haven't loaded your data yet, make sure to load it using pd.read_csv() or a similar method

# 1. Calculate Customer Value
customer_df = df.groupby('Customer ID').agg({
    'Order ID': 'nunique',     # Purchase frequency
    'Sales': ['mean', 'sum']   # Average purchase value and total sales
}).reset_index()

customer_df.columns = ['Customer ID', 'Purchase Frequency', 'Average Purchase Value', 'Total Sales']

# 2. Explore Customer Segments
# Creating segments based on Total Sales
bins = [0, 500, 1000, 5000, float('inf')]
labels = ['Low Value', 'Medium Value', 'High Value', 'Very High Value']
customer_df['Customer Segment'] = pd.cut(customer_df['Total Sales'], bins=bins, labels=labels)

# 3. Identify Patterns and Trends
# Analyzing purchasing behavior over time
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Year'] = df['Order Date'].dt.year

# Total Sales per Year
sales_trends = df.groupby('Year')['Sales'].sum()

# Plotting Sales Trends
plt.figure(figsize=(10, 6))
sns.lineplot(x=sales_trends.index, y=sales_trends.values)
plt.title('Total Sales Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()

# 4. Product Analysis
# Analyzing top-selling products
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(10)

# Plotting Top Products
plt.figure(figsize=(12, 8))
top_products.plot(kind='bar', color='skyblue')
plt.title('Top 10 Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.show()

# 5. Customer Retention
# Calculating and analyzing customer retention rates
retention_df = df.groupby(['Year', 'Customer ID']).size().reset_index(name='Order Count')
retention_rate = retention_df.groupby('Year')['Order Count'].value_counts(normalize=True).unstack()

# Plotting Customer Retention Rate
plt.figure(figsize=(12, 8))
sns.heatmap(retention_rate, cmap='Blues', annot=True, fmt='.1%')
plt.title('Customer Retention Rate Over Time')
plt.xlabel('Order Count')
plt.ylabel('Year')
plt.show()
```
