```python
#Ex.1
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a pandas DataFrame with random values for x and y
np.random.seed(0)
data = {
    'x': np.random.rand(20),
    'y': np.random.rand(20)
}
df = pd.DataFrame(data)

sns.scatterplot(x='x', y='y', data=df)

plt.xlabel('X-Value')
plt.ylabel('Y-Value')
plt.title('Basic Scatter Plot')

plt.show()
```

```python
#Ex.2
categories = np.random.choice(['A', 'B', 'C'], 30)
data = {'category': categories}
df = pd.DataFrame(data)

category_counts = df['category'].value_counts().reset_index() #count the frequency of each category
category_counts.columns = ['Category', 'Frequency']

plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Frequency', data=category_counts)
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Category Frequency')
plt.show()
```

```python
#Ex.3
x = np.linspace(0, 10, 100) #sequence of numbers for x
y1 = np.sin(x)
y2 = np.cos(x)

# Create a DataFrame with x, y1, y2
data = {'x': x, 'y1': y1, 'y2': y2}
df = pd.DataFrame(data)

# Plot y1 and y2 against x using Seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(x='x', y='y1', data=df, color = 'red', label='sin(x)')
sns.lineplot(x='x', y='y2', data=df, color = 'blue', label='cos(x)')

# Add labels and title
plt.xlabel('X-Value')
plt.ylabel('Y-Value')
plt.title('Comparison of Two Functions')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
