```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'/Electronics.csv')
df.head()
```

```python
products = df['Product Purchased']
counts = products.value_counts()
counts
```

```python
explosions = [0] * len(counts)
explosions
```

```python
max_label = counts.idxmax() # maximum values
```

```python
explosions[counts.index.to_list().index(max_label)] = 0.2
```

```python
explosions
```

```python
fig, ax = plt.subplots()
plt.pie(x = counts,labels = counts.index, autopct='%1.1f%%', explode = explosions)
plt.title('The amount of each product purchased')
plt.show()
```
