Data visualization matters because it helps people understand complex information quickly and spot trends or patterns. The functions of data visualization are : Clarity and Understanding , Pattern Identification,Communication and Storytelling.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'/content/sales_data.csv')
df.head()
```

```python
df['quantity'].sum() #the total quantity of products sold by the company.
```

```python
category_revenue = df.groupby('category')['revenue'].sum()
category_highest_revenue = category_revenue.idxmax()
highest_revenue = category_revenue.max()

print(f'Catagories are: {category_revenue}')
print(f'Catagory highest revenue is: {category_highest_revenue}')
print(f'It generates revenue : {highest_revenue}')
```

```python
average_revenue_per_sale = df['revenue'].mean()
print(average_revenue_per_sale)
```

```python
df['date'].value_counts
```

```python
df['date'] = pd.to_datetime(df['date'])  # Convert 'date' column to datetime format
df['quarter'] = df['date'].dt.to_period('Q') # Extract quarter information

quarterly_revenue = df.groupby('quarter')['revenue'].sum()
quarterly_revenue.plot(kind='bar', color='skyblue')
plt.xlabel('Quarter')
plt.ylabel('Total Revenue')
plt.title('Total Revenue Generated in Each Quarter')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add horizontal gridlines
plt.show()
```

```python
import numpy as np
x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y, label='y = x^2', color='pink')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2')

plt.legend()
plt.grid(True)
plt.show()
```

```python
products = ['A', 'B', 'C', 'D']
sales_values = [15, 30, 45, 20]

plt.bar(products, sales_values, color='darkblue')

plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Weekly Sales of Different Products')

plt.grid(axis='y')
plt.show()
```

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
percentages = [40, 30, 20, 10]
colors = ['gold', 'lightgreen', 'lightcoral', 'lightskyblue']

plt.pie(percentages, labels=fruits, colors=colors, autopct='%1.1f%%', startangle=140)
plt.legend(title='Fruits', loc='best')
plt.title('Favorite Fruits')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```
