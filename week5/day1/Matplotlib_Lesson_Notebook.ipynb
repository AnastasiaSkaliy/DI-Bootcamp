# Getting Started With Matplotlib

Figure: Think of the Figure as the overall window or page where your plot (or plots) will appear. You can create multiple independent figures.

# Pie Charts With Matplotlib


## Creating A Basic Pie Chart
Let’s start with a basic pie chart. Suppose we have data representing the market share of different smartphone brands.

```python
import matplotlib.pyplot as plt

# Data
brands = ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Others']
market_share = [40, 30, 15, 10, 5]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(market_share, labels=brands, autopct='%1.1f%%')
plt.title('Smartphone Market Share 2023')
plt.show()
```

In this example, market_share holds the market shares of each brand, and brands contains the corresponding brand names. The plt.pie function creates the pie chart, with autopct parameter used to display the percentage share of each slice.


## Customizing The Pie Chart
Exploding Slices: You can “explode” one or more slices from the pie to highlight them.

```python
explode = (0.05, 0, 0, 0, 0)  # Only explode the first slice (Apple)

plt.pie(market_share, labels=brands, autopct='%1.1f%%', explode=explode)
plt.title('Smartphone Market Share 2023')
plt.show()
```

Changing Colors: You can change the color of each slice.


```python
colors = ['gold', 'lightblue', 'lightgreen', 'orange', 'purple']

plt.pie(market_share, labels=brands, autopct='%1.1f%%', colors=colors)
plt.title('Smartphone Market Share 2023')
plt.show()
```

Adding Shadow: Add a shadow for a 3D effect.

```python
plt.pie(market_share, labels=brands, autopct='%1.1f%%', shadow=True)
plt.title('Smartphone Market Share 2023')
plt.show()
```

# Box Plots In Matplotlib


## Creating Box Plots With Matplotlib


A boxplot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays summary statistics such as the median, quartiles, and potential outliers in a concise and easy-to-understand manner. A boxplot is particularly useful for visualizing the spread and central tendency of data, as well as identifying any extreme values (outliers).

Here's how to interpret a typical boxplot:

Box: The box in the middle of the plot represents the interquartile range (IQR), which spans from the first quartile (Q1) to the third quartile (Q3). The length of the box shows the spread of the middle 50% of the data.

Line inside the box: This line represents the median (Q2), which is the middle value when the data is sorted.

Whiskers: The whiskers extend from the edges of the box to the minimum and maximum values within a specified range. The default for matplotlib is for the whiskers to be 1.5 times the length of the interquartile range.

Outliers: Individual data points that fall outside the whiskers are considered outliers and are typically marked as individual points. They are often indicated by dots or asterisks.

One example for using a boxplot is to visualize data spread: Boxplots help visualize the spread of data, including the variability, skewness, and central tendency.

In education, you can create boxplots to display the distribution of test scores for a class of students, highlighting the median score, quartiles, and potential outliers.

## Basic Box Plot


Here’s how you can create a simple box plot:

## Customizing Box Plots


```python
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 30, 200)

data = [data1, data2, data3]

fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)

# Creating axes instance and the box plot
bp = ax.boxplot(data, patch_artist = True, notch = True, vert = 0)

# Changing color and line width of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color ='blue', linewidth = 1.5, linestyle =":")

# Changing color and line width of the caps
for cap in bp['caps']:
    cap.set(color ='red', linewidth = 2)

# Changing color and linewidth of the medians
for median in bp['medians']:
    median.set(color ='green', linewidth = 3)

# Changing style of fliers
for flier in bp['fliers']:
    flier.set(marker ='D', color ='#e7298a', alpha = 0.5)

# Adding title
plt.title('Customized Box Plot')
plt.show()
```

In this example, we create a more complex box plot with three different datasets. We customize the whiskers, caps, medians, and fliers of the box plot for better visualization and clarity.


## Interpreting Box Plots (Reminder)
A box plot displays the distribution of data based on a five-number summary:


Minimum: The lowest data point excluding any outliers.
Q1 (First Quartile): The median of the first half of the dataset.
Median (Second Quartile, Q2): The middle value of the dataset.
Q3 (Third Quartile): The median of the second half of the dataset.
Maximum: The highest data point excluding any outliers.
Outliers are indicated as individual points that lie outside the overall pattern of the distribution.

## Interactive And Animated Plots In Matplotlib
In this section of the lesson, you will learn how to elevate your data visualizations by creating interactive and animated plots using Matplotlib. These techniques can be especially useful for making dynamic visualizations that change in response to user input or over time.


Key Functions:


mpl_connect: For connecting event handlers to the plot.
Widgets like Slider, Button, and RadioButtons for user inputs.


```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Initial data
t = np.linspace(0, 1, 1000)
initial_freq = 5
y = np.sin(2 * np.pi * initial_freq * t)

fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.25)
l, = ax.plot(t, y)

# Slider
axfreq = plt.axes([0.1, 0.1, 0.65, 0.03])
freq_slider = Slider(axfreq, 'Freq', 0.1, 30.0, valinit=initial_freq)

# Update function
def update(val):
    freq = freq_slider.val
    l.set_ydata(np.sin(2 * np.pi * freq * t))
    fig.canvas.draw_idle()

# Connect the update function to the slider
freq_slider.on_changed(update)

plt.show()
```

## Animated Plots
The objective is to create plots that change over time, useful for showing trends, progress, or dynamic systems.


Key Functions:


FuncAnimation: From the matplotlib.animation module for creating animations.
Example: Creating a basic animated line plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r-', animated=True)

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)

plt.show()
```

In this example, a sine wave is animated to progress over time, providing a dynamic visualization of the function.
